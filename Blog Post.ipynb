### What is Fantasy Football?

Fantasy Football is a game in which a group of people, traditionally around 10 people per fantasy league, act as 'general managers' or 'owners' of their own National Football League Team. They draft 15 players, ranging all positions (as listed below), in a 15 Round  draft (like that in which collegiate players are drafting into the NFL-National Football League each year). All NFL players are available for selection and it is up to the 'general manager'/ the person drafting to select who they want on their team. This is very much a draft of strategy. Everyone has different strategy but there is a general scheme that should be loosely followed. 

The positions are:
- QB = Quarterback
- RB = Running back
- WR = Wide Receiver 
- TE = Tight End
- FLEX = a position in fantasy line-up where you can put an extra WR, RB, or TE
- K = Kicker
- D/ST = Defense/Special Teams

### Diving into the numbers: Players salaries and more

Does the top ranked fantasy players and top average salaries correlate with each other. This is interesting. As my fantasy football story has panned out it is evident that quarterbacks are most frequently the #1 fantasy football players each season. It also happens to be that the Top-20 salaries for NFL players (as of 2020) contains 17 QB's.

So.. not only are quarterbacks the most likely to score the most fantasy points each year, this also correlates with player salaries. 6 out of the 10 #1 ranked fantasy players also have one of the top salaries in the NFL. Again player salary is completely dominated by the Quarterback position.

Although a Quarterback often holds the #1 player rank for fantasy football each season AND is among one of the highest paid players in the League...a quarterback is often not your 1st, 2nd, or even 3rd pick in a fantasy draft. It is important to note that although they are not the first pick in the draft, they are the single most important player and biggest point generator on your fantasy team. 

Chart by Abby Bosco; Source: https://www.spotrac.com/nfl/rankings/average/

Top-20 salaries in 2020 consists of 17 Quarterbacks, 2 Defensive Ends (which isn't its own position in fantasy so you would draft the defensive unit as a whole, not just them as an individual player) and 1 WR. 

Patrick Mahomes, Quarterback of the KC Cheifs, was the #1 ranked player AND MVP in 2019 and has the highest average salary in the league in 2020. Deshaun Watson and Russell Wilson also have 2 of the highest salaries in the league and happen to be very big point generators for fantasy football. 

Takeaway: Look at salary on a positinoal basis. The highest paying players within each position are going to be the best at what they do. It is potentially unhelpful in drafting your team to only look at the Top salaries amongst all players because Quarterbacks are by far the highest paid players in the league. Looking into a salary breakdown by position can really help you find who your target draft picks should be.

### Professional Draft Tips:
* What do the NFL professionals say about how you should draft your fantasy team?
* How does my analysis align with this and how is it able to explain this?

#### First lets lay-out the fantasy football line-up:
    1 Quarterback (QB)
    2 Running Backs (RB)
    2 Wide Receivers (WR)
    1 Flex (choose from RB/WR/TEs)
    1 Tight End (TE)
    1 Placekicker (K)
    1 Team Defense/Special Teams (D/ST)

- there are also 6 Bench (BN) players. The players in your bench DO NOT count for the points total of your team during your match-up.

#### According to NFL Professionals, this is how they suggest you should draft your fantasy team for success:

#### TIP #1: In Round 1 you should draft a RB
   This was interesting because if you are going based off of who generates the most amount of points for your team it should be obvious that you should pick a Quarterback. But heres the thing, there are 9 to 10 quarterbacks each year that make it into the Top-10 of the ranks (with the range of point totals not being very drastic). There are typically only 10 teams per league. That being said, although you might not get the best quarterback by waiting till a much later round, there is a very good change of you drafting a quarterback that is still a top-10 or top-20 point generator. Good, point generating QB's aren't scare. BUT.. RB's are for a number of reasons. And, it is so important to draft at least 1 really good one because you need 2 in your line-up each week. Problem #1 with running backs is that they do not play every play and often teams rotate between 2 or 3 of them. If you can find a team that has a distinct and set running back that they go to all the time, you need to get them early. Second problem is there is a huge drop-off with RB's. There are a few that make the top charts, but only a few. If you wait too long to get a running back, you're going to be stuck with ones who simply do not generate enough points for you to succeed. So.. tip #1 is... get a good/go to running back AND draft them EARLY.
    
#### TIP #2: In Round 2 you should draft a WR.
   Wide receiver similar to RB except not as important because  typically RBs generate more total points. But you need 2 of them in your lineup and there is also a pretty big drop-off in total points between the 10-15 best in the league and the rest. There are not enough really big point generating WR's for each team to have 2 of them. Again, like RBs there are multiple strings of WRs, you want to make sure you are able to draft at least 2 first string WRs that are high point generating. If you wait too long to draft a WR you might get stuck with a player that is not their teams "go-to" guy and therefore they don't get as many reps in the games and will therefore not generate many points. Only 5 wide receivers made it into the Top-20 in any of the past 10 years (as shown in the graph below). This emphasizes the importance of getting at lest one good WR because there are a few good ones and then there is a drop off.

#### TIP #3: Round 3-6 you should aim to draft either a RB or WR (make sure you are picking even #'s of each):
   Get them early. Get at least 2 really good "go-to" first string players. They go fast and you need to get them while they can. The drop-off of these positions is much higher than that of Quarterbacks

#### TIP #4: Round 7 draft a TE
   TEs are often not some of the bigger point scorers. I would describe them as they background player. They don't appear frequently in the Top of the ranks. You only need to play 1 of them in your line-up so its not essential to focus on getting more than 1 reallyy good one. As shown in the graphs above and below, only 1 TEs, over the past 10 years, has appeared even in the Top-20 fantasy player ranks. They are not going to generate a ton of points like the RBs and WRs but it is important to draft at least 1 first string TE being they get substantially more plays and reps in the game than 2nd and 3rd string TEs.

#### TIP #5: Around Round 8-9 you want to draft a QB
   If you wait past round 8 to draft a QB that just means you most likely will not get the best, top-tier, high- caliber QBs that are often round in the top 5 spots of the ranks. Many of the really good quarterbacks will be around during round 8... although there is a chance someone snags the #1 slightly earlier. 

#### TIP #6: Rounds 9-12 is more about drafting the best available player at either the QB, WR, RB, or TE position
   These players are most likely going to be on your bench so it is not super important. It is important to draft the best player available that is one of these positions that it projected to generate the most points. My biggest piece of advice here is look to last years statistics to determine which player is the better choice. There is not much discrepancy in Top players. Meaning, if they are in the Top-50 one year, it is likely they will generate  decent amount of points the next year as well. This is an opportunity to get the best available bench players possible. These players will most likely remain on your bench unless one of your starters get injured, one of your starters has a bye week, or a bench player is simply having a breakout season and was a great pick late on in the draft (exceeded expectations).

#### TIP #7: Round 13 draft a kicker.
   Kicker's often don't generate many points and there are a lot of them who are about equal with the exception of the 1-2 kickers who are just exceptional. That being said they don't play much and often do not get the opportunity to generate a great deal of points. Throughout the past 10 years, there has only been 2 occassions in which a Kicker appeared in the Top-50 (they never appeared in the Top-10 or Top-20) and the highest rank a Kicker has held is #40. 

#### Tip #8: Round 14-15 draft a Defense/Special Teams (D/ST) or K.
   Defense's are very up-and-down on a point generating term completely depending on who they are playing. When they are playing a poor offensive team, many D/ST rack up points, but on a week where they are playing a well oiled offense, they often will only generate a few points (sometimes they will generate '0' or even a negative number). It is ideal to get one of the top-5 ranked Defenses but this is a position that is going to be available still at the very end of the draft.

### The Complete Breakdown and Explanation to it All:

Chart by Abby Bosco; Source: https://fantasydata.com/nfl/fantasy-football-leaders


This is the graph that says it all. It is a breakdown of the average total points of each position over the past 10 years vs. how many players of that position are listed as a Top-50 fantasy point scorer. 

Quarterbacks are the Kings of the game. They score the most points, the win the most MVP awards, and they make the most money. So...my question this whole time is why wouldn't you pick a QB in the first round if they have completely dominated the fantasy football ranks for the past 10 years. The answer is in this graph here itself..Yes, quarterbacks generate the most amount of total points each year BUT they are also the position with the most amount of players in the Top-50. They're a hot commodity but there is a high supply (just like the supply-and-demand chain). So, don't waste your number 1st, 2nd, or 3rd draft pick on these guys because even if someone else in the league does, theres enough high-caliber quarterbacks in the league that you are almost guaranteed to end up with one. 

Running backs. Why should they be selected by all league 'general managers' Round 1? As depicted in the graph above, the Running Back position is the second highest point generating position in the league. The big thing to take into account here is that there are only half as many RB's that generate enough points to make the Top-50 cut-off than there are quarterbacks. They are also of high demand not only because they generate a decent amount of points but also because you need 2 of them in your starting line-up. Very high demand with a slightly low availability. This is why you have to get the best possible RB that you can in the first round.

Drafting a wide receiver early has a similar reasoning behind it. They don't generate as many points as a RB, but there are also few WR's who appear in the Top-50 of the ranks. It is important that you get one of them early because, like RB's.. 2 WR's start in your line-up. It is essential to score at least 1 of the top WRs in order for your team to succeed. To do this, you need to draft them early or another 'general manager' will beat you to it.

TE, K's, and D/ST. All very low supply in the top-50 ranks over the past 10 years. They're not the biggest point generator on any single fantasy team and you only need one of each in your line-up. Don't make this your focus but also don't push off getting a TE until the very last round. You want to make sure you pick at least 1 TE who is their specific NFL teams' first string or else your TE will not get sufficient even plays in the game to impact your team enough to succeed. With K's and DST you usually only select one per team and there are plenty first strings to go around. Definitely save these for your last 2-3 picks (there aren't many in the top of the ranks and if they even did make the top-50 they made the very tail end of it). 

I can't promise you will come in first place but if you follow this total guide to fantasy football I can assure you, you are doing it right. Pick smart, do your research, and take some risks on certain players (they might surprise you and be the difference maker).


### Last thing, a subtle reminder: Where to look for the trends to help draft your team...
Here is a list of my tips on where to look to find trends to help you select the best team possible.

#1: Look to the Top-50 fantasy lists from the previous years...if a player is on multiple...take a risk he is probably going to be on it again (although I wouldn't consider that much of a risk!)

#2: Look to average salaries. On a positional basis, which players are being paid the most. The more the player is paid (not 100% of the time but MOST of the time) the more fantasy points they generate. Although it is not an intended correlation because that is not what NFL teams look at (fantasy points) when they give a player a contract but it is no coincidence that the best players in the league are getting paid the most AND consistently total some of the most fantasy points in the League.

#3: SuperBowl contenders does not correlate with top fantasy players. I WOULD NOT recommend specifically looking at just players who were on a SuperBowl contending team the year before. There is no evidence that proves that these SuperBowl teams are the ones with players putting up a high fantasy points total. But... don't get me wrong, there are some instances where the shining star of the fantasy football league is also the SuperBowl MVP (like Patrick Mahomes in 2019 -- he did it all).

#4: There are a few teams who have had more players appear in the Top-50 over the past 10 years than others. But, this isn't extremely reliable as it can change year-by-year and it is really dependent on the player themself and what they are capable of producing on the field. It is important to keep in mind that the better teams in the NFL are going to have better chances of producing more points. For example, picking a WR on a team with an awful quarterback is not smart because although that WR is able to make a play the QB is terrible at getting him the ball.. therefore he is not going to generate the points that you need. The team the player plays for is by no means the most important factor to consider when drafting your team BUT.. it is something that should not be overlooked and can help you get the upperhand in some cases.

#5: When thinking about your draft order REMEMBER...It all comes down to supply and demand. Strategize when selected when you want to draft who because remember these few key things I've learned through developing this story:
    * QB's are the most point generating players but there are enough really high point generating QB's to go around (the drop off of the top-10 QB's is not substancial and there are only 10 teams in each fantasy league). Although drafting the #1 QB is ideal, it is not going to make or break your team.
    
    * Draft a RB early. They are high point generators but there are only a few really notable RB's that will drastically impact your team. Get one of them and you are destined to succeed.

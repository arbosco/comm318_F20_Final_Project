### II. In which year was it the hardest to break into the Top-50?
* Analyzing the mean total points for each year

```python
# 1. create a list of data frames 
list_of_dfs = [fantasystats_2010_df, 
               fantasystats_2011_df, 
               fantasystats_2012_df,
              fantasystats_2013_df,
              fantasystats_2014_df,
              fantasystats_2015_df,
              fantasystats_2016_df,
              fantasystats_2017_df,
              fantasystats_2018_df,
              fantasystats_2019_df]


combined_data=[]

# 2. loop over each data frame and select top 50 (the whole data frame) and add a year column
for year, df in enumerate(list_of_dfs, 2010):
    df['YEAR']=year
    combined_data.append(df)
    
    
# 3. create a combined data frame
fantasystats_df = pd.concat(combined_data)
```

```python
# compare the minimum values to break the Top-50 for each year
# using the Data Frame created above 
fantasystats_df.groupby('YEAR')['FPTS'].min()
```

#### Analysis:
The minimum points needed to break the Top-50 list 164.10 total points. This was the the 50th ranked players points total ('FPTS') for the 2017 regular season. 

Since the minimum value was the highest in the year 2013, this tells us that in this year it was hardest to break into the Top-50.

```python
# find the maximum value (the #1 ranked player)
fantasystats_df.groupby('YEAR')['FPTS'].max()
```

#### Analysis:
* The highest amount of points scored was in 2018. This means the hardest year to be crowned the #1 ranked fantasy football player was in 2018. You needed the most amount of points to be crowned #1 in this year.

```python
# compare the average scores for each year 
fantasystats_df.groupby('YEAR')['FPTS'].mean()
```

#### Analysis:
The average points needed to be in the Top-50 was the highest in 2018. This indicates that the highest number of players generated high point totals during the 2018 regular season.

#### Additional Analysis: 
These are the statistics it takes each year to make it into the Top-50 list for most fantasy generated points:
- Ranges:
    * 2010: 326.00 - 169.50 
    * 2011: 397.18 - 171.90
    * 2012: 345.58 - 182.00
    * 2013: 409.98 - 187.40
    * 2014: 354.14 - 176.12 
    * 2015: 389.08 - 171.50
    * 2016: 380.02 - 171.70
    * 2017: 347.92 - 164.10
    * 2018: 417.08 - 181.40
    * 2019: 415.68 - 176.50


## Answer: 
The hardest year, meaning the year in which you had to score the highest amount of points to be in the Top-50, was 2018. The #1 year scored the most amount of points in this year AND the average number of total points needed to be in the Top-50 was also highest in 2018.

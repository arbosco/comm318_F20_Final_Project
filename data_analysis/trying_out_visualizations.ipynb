### Analysis: 4 players have appeared in all 10 Top-50 tables:
- All 4 are quarterbacks 
- Philip Rivers, Drew Brees, Matt Ryan, and Tom Brady

```python
#create visualization to see progression of players who have appeared in all 10 of the lists 
# Philip Rivers, Matt Ryan, Aaron Rodgers, and Drew Brees

#step 1: create data frame of each of their 10 appearances
Rivers_10 = fantasystats_df[(fantasystats_df['NAME'] == 'Philip Rivers')]
Brady_10 = fantasystats_df[(fantasystats_df['NAME'] == 'Tom Brady')]
Ryan_10 = fantasystats_df[(fantasystats_df['NAME'] == 'Matt Ryan')]
Brees_10 = fantasystats_df[(fantasystats_df['NAME'] == 'Drew Brees')]

#step 2: create combines data set
list_of_10_dfs = [Rivers_10,
                 Brady_10,
                 Ryan_10,
                 Brees_10]
combined_10_data =[]

for year, df in enumerate(list_of_10_dfs):
    df=df.head(10)
    df['YEAR']=year
    combined_10_data.append(df)
    
fantasy_toprepeats = pd.concat(combined_10_data)
```

```python
fantasy_toprepeats
```

```python
fantasy_toprepeats_df=fantasy_toprepeats[['NAME', 'FPTS']]
fantasy_toprepeats_df.plot(x='NAME', y='FPTS', kind='bar', figsize=(12,4))
```

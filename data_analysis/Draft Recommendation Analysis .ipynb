### Fantasy Football Line-Up:

Each team in a fantasy league plays 9 players (as outlined above). The bench players points will not count in the match-up total and therefore will not contribute to the outcome of the match. Each week the 'general managers' need to set their fantasy line-up and it must consist of:



* 1 Quarterback (QB)
* 2 Running Backs (RB)
* 2 Wide Receivers (WR)
* 1 Flex (choose from RB/WR/TEs)
* 1 Tight End (TE)
* 1 Placekicker (K)
* 1 Team Defense/Special Teams (D/ST)
* 6 Bench (BN)

### Quarterback:
- recommended to be picked somewhere between rounds 8-12

Initially I was VERY surprised by this. From my data exploration & analysis I found that the most common position to appear in the Top-50, Top-10, Top-1, League MVP, and Top-20 salary lists was overwhelmingly Quarterback. But... the thing with quarterbacks as seen in the charts I analyzed, there are MANY of them who score A LOT (9/10 of the top-10 are usually quarterbacks). So... there are PLENTY good point generators available to draft.

When it comes to other positions that is not always the case. There are a few of each position scattered throughout the Top-50 BUT.. only a few. And...a fantasy football line-up consists of multiple RBs, WRs, TEs, and then there is a FLEX position (where you can play any of those 3 positions). You not only need multiple of these position in your line-up BUT...there are only few players in these positions who generate enough points to break into the Top-50. People need to select these players earlier on because there are less "REALLY GOOD" ones and more spots for them in the line-up at the same time.

This is why, as you will see below, it is recommended these 3 positions be selected earliest on.


follow-up: why are Kickers and Defenses recommended in the last rounds?
- barely any break the Top-50
- you only need 1 of each in the line-up
- clear from not breaking the Top-50 -- they generally are the least point generating positions unless you draft the best D or the best K. (is that worth it in sacrificing a higher up spot for another position)
